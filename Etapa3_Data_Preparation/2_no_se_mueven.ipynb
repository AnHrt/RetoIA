Codigo para limpieza de los usuarios que nunca se mueven

```python
# Librerias
import os # Sistema operativo
import pandas as pd # Lectura de Dataframe
import numpy as np # Calculos
```

```python
# Importamos el dataset que tiene el cálculo de las distancias
df = pd.read_csv('../dataset/20210101_RM_jumpDist_sinSumar.csv')
df.head()
```

```python
# Agrupamos las distancias por PHONE_ID y sumamos todas las distancias recorridas por cada usuario
df = df.groupby('PHONE_ID', as_index=False)['jumpDist'].sum()
df
```

```python
# Creamos una columna de borrado lógico, True - cuando la suma de la distancia es 0 (no se mueve)
df['borrar'] = np.where(df['jumpDist'] != 0, False, True)
df
```

```python
# Contamos los datos, False: datos que no se borran / True: atos que se deben borrar
df.borrar.value_counts()
```

```python
# Optimizamos el dataset eliminando la columna de la distancia
# Sólo mantendremos PHONE_ID y el borrado lógico
df = df.drop(columns=['jumpDist'])
df
```

```python
# Cargar dataset con distancia, tiempo y velocidad
# Actualización 06/11/2022, cargamos el dataset que le limpiamos 
# las velocidades mayores a 150 km/h, hubo un error pero se soluciona
# facilmente volviendo a realizar el borrado de los que no se mueven
df2 = pd.read_csv('./complete/20210101_RM_jumpTime_jumpSpeed_clean.csv')
df2
```

```python
# Eliminar columnas (ya no se eliminan porque las necesitamos para las velocidades extremas)
# df2 = df2.drop(columns=['timestamp', 'bts_id', 'lat', 'lon'])
```

```python
# Juntamos el dataset del borrado lógico con el dataset de distancia, tiempo y velocidad
# Actualización 06/11/2022: Con el dataset de las velocidades > 150km/h borradas
df3 = df.merge(df2, on = 'PHONE_ID', how = 'left')
df3
```

```python
# Verificamos que al hacer merge, se haya hecho un fill de true y false respecto al merge con PHONE_ID
df3['borrar'].isna().sum()
```

```python
# Sólo para asegurarnos contamos los registros que se van a eliminar (True) y los que se mantendrán (False)
df3.borrar.value_counts()
```

```python
# Eliminar los usuarios que no se mueven con la columna del borrado lógico
df3 = df3.drop(df3[df3.borrar == True].index)
df3
```

```python
# Quitamos la columna del borrado lógico
dfFinal = df3.drop(columns=['borrar'])
dfFinal
```

```python
# Guardamos el dataset con los usuarios que sí son activos
# dfFinal.to_csv('../dataset/20210101_SiSeMueven.csv', index=False)
# Actualización 06/11/2022, ahora hacemos que este sea el verdadero final
dfFinal.to_csv('./complete/20210101_RM_final_06_11_2022.csv', index=False)
```
