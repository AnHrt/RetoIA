# Codigo para calcular distancias, tiempo y velocidad de los viajes

```python
#Librerias 
import os            #Sistema operativo
import pandas as pd  #Lectura de Dataframe
import numpy as np   #Calculos
```

```python
#Importar dataset
df = pd.read_csv('../dataset/20210101_RM.csv')
```

```python
trazas = df.copy()
trazas
```

```python
#Acomodar datos por usuario y por fecha 
trazas = trazas.sort_values(by=['PHONE_ID','timestamp'])
trazas
```

```python
trazas.info()
```

```python
#Calcular distancia apartir de la longitud y latitud
def distance_d(LaA, LaB, LoA, LoB):
    #LaA & LoA - primer punto
    #LaB & LoB - punto final
    
# The function "radians" is found in the numpy module, It's also used to convert radians to degrees.  
    LoA = np.radians(LoA)  
    LoB = np.radians(LoB)  
    LaA= np.radians(LaA)  
    LaB = np.radians(LaB) 
    
    # The "Haversine formula" is used.
    D_Lo = LoB - LoA 
    D_La = LaB - LaA 
    P = np.sin(D_La / 2)**2 + np.cos(LaA) * np.cos(LaB) * np.sin(D_Lo / 2)**2  
    #trazas.sort_values(by=['PHONE_ID','timestamp'])
    Q = 2 * np.arcsin(np.sqrt(P))
    # The earth's radius in kilometers.
    R_km = 6378  
    # Then we'll compute the outcome.
    return(Q * R_km)
```

```python
trazas['lat'].apply(lambda x: float(x))
trazas['lon'].apply(lambda x: float(x))
trazas['jumpDist'] = distance_d(trazas.groupby(['PHONE_ID'])['lat'].shift(1), trazas['lat'], trazas.groupby(['PHONE_ID'])['lon'].shift(1), trazas['lon'])
```

```python
trazas
```

```python
#Elimina las columnas para solo quedarse con el telefono y la distancia 
#Dataset optimizado para quitar a los usuarios que no se mueven
trazas_final = trazas.drop(columns=['timestamp', 'bts_id', 'lat', 'lon'])
trazas_final
```

```python
trazas_final.to_csv('../dataset/20210101_RM_jumpDist_sinSumar.csv', index=False)
```

```python
#Modifica la fecha de "objeto" a "datetime"
trazas["timestamp"] = pd.to_datetime(df.timestamp)
```

```python
#Calculo del tiempo en hora
trazas['jumpTime'] = (trazas['timestamp'] - trazas.groupby(['PHONE_ID'])['timestamp'].shift(1)) / pd.Timedelta(hours=1)
```

```python
trazas['jumpTime']
```

```python
#Calculo de la velocidad, con tiempo y distancia KM/HR
trazas['jumpSpeed'] = trazas['jumpDist'] / trazas['jumpTime']
trazas
```

```python
trazas_final_velocidad = trazas
```

```python
#Coloca 0 en los datos que contienen NA (los cuales son los registros iniciales)
trazas_final_velocidad["jumpDist"] = trazas_final_velocidad["jumpDist"].fillna(0)
trazas_final_velocidad["jumpTime"] = trazas_final_velocidad["jumpTime"].fillna(0)
trazas_final_velocidad["jumpSpeed"] = trazas_final_velocidad["jumpSpeed"].fillna(0)
trazas_final_velocidad
```

```python
#Generar csv con el dataset
trazas_final_velocidad.to_csv('../dataset/20210101_RM_jumpTime_jumpSpeed.csv', index=False)
```
